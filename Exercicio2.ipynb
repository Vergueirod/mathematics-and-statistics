### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = 2x + 3
# y = 2x + 3

x = np.linspace(5, 10, 100)
def f(x):
    return (2 * x) + 3

y = f(x)
y
```

```python
# Cria o gráfico
plt.figure(figsize = (6, 4))
plt.plot(x, y, label = 'f(x) = 2x + 3')

# Adiciona um título e rótulos aos eixos
plt.title('Gráfico da Função linear')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona uma legenda
plt.legend()

# Mostra o gráfico
plt.grid(True)
plt.show()
```

### Exercício 2: Gráfico de uma Função Quadrática

Desenhe o gráfico da função y = x^2 - 4x + 4. Crie uma massa de dados de exemplo usando np.linspace().

```python
x = np.linspace(-10, 10, 400)

def f(x):
    return (x ** 2) - (4 * x) + 4

y = f(x)
y
```

```python
# Cria o gráfico
plt.figure(figsize = (6, 4))
plt.plot(x, y, label = 'f(x) = x^2 - 4x + 4')

# Adiciona um título e rótulos aos eixos
plt.title('Gráfico da Função Quadrática')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona uma legenda
plt.legend()

# Mostra o gráfico
plt.grid(True)
plt.show()
```

### Exercício 3: Gráfico de Uma Função Senoidal

Crie um gráfico da função y = sin(x) para valores de x no intervalo de 0 a 2π.

```python
random_x = np.linspace(0, 2*np.pi , 100)

seno = np.sin(random_x)
seno

```

### Exercício 4: Gráfico de Uma Função Exponencial

Crie um gráfico para a função y = e^x para valores de x no intervalo de -2 a 2.

```python
# y = 2.71828 ** x

e = 2.71828

random_x = np.linspace(-2, 2 , 100)

y = e ** random_x
y
```

```python
# Cria o gráfico
plt.figure(figsize = (6, 4))
plt.plot(random_x, y, label = 'f(x) = e^x')

# Adiciona um título e rótulos aos eixos
plt.title('Gráfico da Função Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona uma legenda
plt.legend()

# Mostra o gráfico
plt.grid(True)
plt.show()
```

### Exercício 7: Aplicação de Logaritmos

Utilize logaritmo para calcular o tempo necessário para que um investimento atinja um determinado valor, considerando uma taxa de crescimento contínua.

```python
import numpy as np

def tempo_para_atingir_valor(P, A, r):
    if P <= 0 or A <= 0 or r <= 0:
        raise ValueError("P, A e r devem ser maiores que zero")
        
    t = np.log(A / P) / r
    print(f"Tempo necessário: {t:.2f} unidades de tempo (ex: anos)")
    return t

# Exemplo
tempo_para_atingir_valor(P=1000, A=2000, r=0.05)
```
